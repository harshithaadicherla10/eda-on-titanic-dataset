# **Task-02: Exploratory Data Analysis on Titanic Dataset**

This notebook performs data cleaning and exploratory data analysis (EDA) on the Titanic dataset from Kaggle. It includes handling missing values, visualizing distributions, and analyzing relationships between features such as survival rate across gender, passenger class, and age. The goal is to uncover key patterns and trends that influenced survival outcomes.

```python
import pandas as pd
df = pd.read_csv('train.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)
```

```python
import	seaborn	as	sns
sns.countplot(x='Survived', hue='Sex', data=df)
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
```

```python
df.to_csv('cleaned_titanic.csv', index=False)
```

### **✅ Summary**

- Cleaned the Titanic dataset by handling missing values and removing irrelevant columns.
- Performed exploratory data analysis (EDA) to understand data distributions and relationships.
- Visualized survival rates based on features like gender, passenger class, age, and embarkation point.
- Identified patterns such as higher survival rates among females and first-class passengers.
